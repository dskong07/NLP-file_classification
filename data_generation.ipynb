```python
import os
import pandas as pd
import numpy as np
```

```python
folders = os.listdir('./topics')
print(folders)
d={}
for i in folders:
    temp_path = f'./topics/{i}'
    f=[]
    for subdirs, dirs, files in os.walk(temp_path):
        f.extend(files)
    d[i] = f
```

```python
test = pd.DataFrame(d['Pictures'])
test['category'] = 0
test
```

```python
d.keys()
```

```python
c = 0
dfs = []
for key in d.keys():
    temp = pd.DataFrame(d[key])
    temp['category'] = c
    dfs.append(temp)
    c += 1
dfs
```

```python
data = pd.concat(dfs)
data.to_csv('file_categories.csv')
```
